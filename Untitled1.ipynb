```python
import pandas as pd
path='/content/drive/My Drive/Colab Notebooks/BrandEquity_student_copy.xlsx'
my_data = pd.read_excel(path)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import seaborn as sns
from math import sqrt
```

```python
df = my_data
```

```python
# Get statistics on the dataset.
my_data.describe()
```

```python
df = pd.DataFrame(my_data)
df.head(2)
```

```python
df.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('OBE', axis=1), df['OBE'], test_size=0.2, random_state=42)
```

```python
# Fit a multiple linear regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Make predictions on test data
y_pred = regressor.predict(X_test)

print("Intercept: \n",regressor.intercept_)
print("Coefficients: \n",regressor.coef_)
```

```python
print('R-squared: {:.3f}'.format(r2_score(y_test, y_pred)))
print('MSE: {:.3f}'.format(mean_squared_error(y_test, y_pred)))
print('RMSE: {:.3f}'.format(sqrt(mean_squared_error(y_test, y_pred))))
print('MAE: {:.3f}'.format(mean_absolute_error(y_test, y_pred)))
     
```

```python
sns.scatterplot(y_test,y_pred)
```

```python
error = y_test - y_pred
sns.displot(error, kde=True)
```
